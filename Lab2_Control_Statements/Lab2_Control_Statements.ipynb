break statement

Terminates the loop statement and transfers execution to the statement immediately after the loop

```python
for letter in 'Inspiration':
    if letter == 'i':
        pass
        print("This is a pass block")
    print("Letter :", letter)
```

```python
for letter in 'Determination':
  if letter == 'm':
    break
  print( 'Letter: ', letter)
```

```python
for letter in 'Python':
    if letter == 'h':
        break
    print("Letter :", letter)

print("Statement immediate after the for loop")
```

continue statement

Causes the loop to skip the remaining part of its body and immediately retest its condition prior to reiterating

```python
for letter in 'Python':
    if letter == 'h':
        continue
    print("Letter :", letter)

print("Statement immediate after the for loop")
```

pass statement

The pass statement does nothing. It can be used when a statement is required syntactically but the program requires no action.

```python
for letter in 'Python':
    if letter == 'h':
        pass
        print("This is pass block")
    print("Letter :", letter)
```

Comprehensions in Python

Comprehensions in Python provide a short and concise way to construct lists, set, dictionary etc using sequences which have been already defined.

• List Comprehensions
• Dictionary Comprehensions
• Set Comprehensions

List comprehensions

```python
nums = [0, 1, 2, 3, 4]

newnums = []

for i in nums:
    if i % 2 == 0:
        newnums.append(i**3)

print(newnums)
```

```python
nums = [0, 1, 2, 3, 4]

newnums = [i**3 for i in nums if i % 2 == 0]

print(newnums)
```

Dictionary Comprehension

```python
nums = [1, 2, 3, 4, 5]

sq_dict = {}

for i in nums:
    sq_dict[i] = i**2

print(sq_dict)
```

```python
nums = [1, 2, 3, 4, 5]

sq_dict = {i: i**2 for i in nums}

print(sq_dict)
```

```python
nums = [1, 2, 3, 4, 5, 6]

even_sq = {i: i**2 for i in nums if i % 2 == 0}

print(even_sq)
```

Set Comprehension

```python
nums = [1, 2, 2, 3, 4, 4, 5]

sq_set = set()

for i in nums:
    sq_set.add(i**2)

print(sq_set)
```

```python
nums = [1, 2, 2, 3, 4, 4, 5]

sq_set = {i**2 for i in nums}

print(sq_set)
```

```python
nums = [1, 2, 3, 4, 5, 6]

even_set = {i for i in nums if i % 2 == 0}

print(even_set)
```

Python zip() Function

The purpose of zip() is to map the similar index of multiple containers so that they can be used just using as single entity.

zip(*iterators)

Parameters : Python iterables or containers ( list, string etc )

Return Value : Returns a single iterator object, having mapped values from all the containers.

```python
# initializing lists

name = ["Sachin", "Sehwag", "Dravid", "Gambhir"]

roll_no = [10, 44, 19, 5]

score = [99, 91, 95, 90]

todays_score = [10, 13, 75, 98]
```

```python
# Use zip() to map values

mapped = zip(name, roll_no, score, todays_score)

mapped = list(mapped)

print(mapped)
type(mapped)
```
