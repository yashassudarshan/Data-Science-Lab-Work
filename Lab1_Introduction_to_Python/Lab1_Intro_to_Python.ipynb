container types in python
The four core built-in container types are:
list: A mutable, ordered sequence of elements. Lists are indexed by integers and can hold items of different types. They are defined using square brackets (e.g., [1, "a", 2.5]).
tuple: An immutable, ordered sequence of elements. Tuples are similar to lists but cannot be changed after creation. They are defined using parentheses (e.g., (1, "a", 2.5)).
dict (Dictionary): A mutable mapping type that stores data as unordered (ordered since Python 3.6) key-value pairs. Data is accessed using unique keys, not integer indices. They are defined using curly braces (e.g., {"name": "John", "age": 30}).
set: A mutable, unordered collection of unique elements. Sets are useful for operations like union and intersection and for quickly checking membership. They are defined using curly braces (e.g., {1, 2, 3}) or the set() constructor.

Problem 1: Confusion Matrix Metrics (Python as a Calculator)
A binary classification model predicts whether rainfall will cause Flood (Yes/No).
Given:
•	True Positives (TP) = 50
•	False Positives (FP) = 8
•	True Negatives (TN) = 32
•	False Negatives (FN) = 10
Tasks
1.	Calculate total predictions
2.	Compute:
o	Accuracy
o	Precision
o	Recall
o	F1 Score


```python

TP = 50
FP = 8
TN = 32
FN = 10

# Calculation of total predictions
total = TP + FP + TN + FN

# Calculation of accuracy
accuracy = (TP + TN) / total

# Calculation of precision
precision = TP / (TP + FP)

# Calculation of recall
recall = TP / (TP + FN)

# Calculation of F1 score
f1_score = 2 * (precision * recall) / (precision + recall)

# Print results
print("Total Predictions:", total)
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1_score)
```

Problem 2: R² (Coefficient of Determination)
In a regression model predicting river discharge, the following are known:
•	Sum of Squares due to Regression (SSR) = 420
•	Sum of Squares due to Error (SSE) = 180
Tasks
1.	Compute Total Sum of Squares (SST)
2.	Calculate R²
3.	Interpret whether the model explains more than 70% variance


```python

SSR = 420
SSE = 180

# Calculate Total Sum of Squares
SST = SSR + SSE

# Calculate R-squared
R_squared = SSR / SST

# Print results
print("Total Sum of Squares (SST):", SST)
print("R-squared (R²):", R_squared)
```

Interpretation: The model explains 70% of the variance in river discharge.
Since R² is equal to 0.7, the model explains exactly 70%, not more than 70%, of the variance.

Problem 3: Mean Squared Error (MSE) Using Python as Calculator
A model predicts temperature values.
The squared errors for 5 observations are:
(2,4,1,9,4)

Tasks
	Store squared errors in Python
	Calculate Mean Squared Error (MSE)
Print the result


```python
# Store squared errors
e1 = 2
e2 = 4
e3 = 1
e4 = 9
e5 = 4

# Number of observations
n = 5

# Calculate Mean Squared Error
MSE = (e1 + e2 + e3 + e4 + e5) / n

# Print the result
print("Mean Squared Error (MSE):", MSE)
```

Problem 4: Cook’s Distance (Influential Observation)
For a regression model:
	Squared residual (eᵢ²) = 9
	Mean Squared Error (MSE) = 3
	Number of predictors (p) = 3
	Leverage value (hᵢᵢ) = 0.2
	Sample size (n) = 25
Tasks
	Compute Cook’s Distance
	Compare with threshold 4/n
	Decide whether the observation is influential


```python
# Assign given values
ei_squared = 9
MSE = 3
p = 3
hii = 0.2
n = 25

# Calculate Cook's Distance
cooks_distance = (ei_squared / (p * MSE)) * (hii / (1 - hii)**2)

# Calculate threshold
threshold = 4 / n

# Print results
print("Cook's Distance:", cooks_distance)
print("Threshold (4/n):", threshold)
```

Cook’s Distance (0.3125) is slightly less than the threshold (0.16), so the observation is not influential.

Problem 5: Simple Linear Regression Prediction (Calculator Style)
A regression model for predicting crop yield is:
y=2.5x+10

Given rainfall value:
	x=12
Tasks
	Predict the value of y
	Print the predicted output
	Explain the meaning of slope and intercept


```python
# Assign slope and intercept values
slope = 2.5
intercept = 10

# Given rainfall value
x = 20

# Predict the output
y_pred = slope * x + intercept

# Print the predicted value
print("Predicted Crop Yield:", y_pred)
```

Explanation:
Slope (2.5): For every 1 unit increase in rainfall, the crop yield increases by 2.5 units.
Intercept (10): When rainfall is zero, the predicted crop yield is 10 units.
